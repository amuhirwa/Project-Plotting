```python
plt.hist("co2", data=fuel_econ, bins=30, edgecolor="black")

plt.title("Histogram of CO2")

plt.xlabel("CO2")
plt.ylabel("Frequency")

plt.savefig("co2_histogram.png", bbox_inches="tight")

plt.show()
```

```python
plt.hist("city", data=fuel_econ, bins=20, edgecolor="black")

plt.title("Histogram of City")

plt.xlabel("City")
plt.ylabel("Frequency")

plt.savefig("city_histogram.png", bbox_inches="tight")

plt.show()
```

```python
plt.hist("ghgScore", data=fuel_econ, bins=10, edgecolor="black")

plt.title("Histogram of GHGScore")

plt.xlabel("GHGScore")
plt.ylabel("Frequency")

plt.savefig("ghg_.png", bbox_inches="tight")

plt.show()
```

```python
# Select numerical columns from the DataFrame
numeric_cols = fuel_econ.select_dtypes(include=[np.number])

# Calculate the correlation matrix for those numerical columns gotten from above
correlation_matrix = numeric_cols.corr()

# Generate a heatmap from the correlation matrix
# The 'diverging_palette' is used to create a custom color palette which matches the palette in the example
sns.heatmap(correlation_matrix, cmap=sns.diverging_palette(20, 220, n=200), fmt='.2f', vmin=-1, vmax=1)

# Add a title to the heatmap
plt.title("Heatmap of Feature Correlations")

# Rotate the x-axis labels by 45 degrees
plt.xticks(rotation=45)

# Save the generated heatmap as a PNG file
plt.savefig("correlation_heatmap.png", bbox_inches="tight")

# Display the heatmap plot
plt.show()
```
